```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def calc(func, start, steps=100):
    xn = start
    w = [start]
    for n in range(steps):
        xn = func(xn)
        w.append(xn)
    return [x for x in range(steps+1)], w

def show(func, start, steps=100):
    x,y = calc(func, start, steps)
    plt.plot(x, y)
    plt.show()
```

```python
def f1(x):
    return 4*x*(1-x)
```

```python
show(f1, 0.249)
```

```python
show(f1, 0.250)
```
